# **Cleaning Some Data**

```
df.drop(['Segment', 'Country' , 'Region'], axis=1, inplace=True)
df.head(10)
```

# **Visuallization**

```
#Scatter 
sns.pairplot(df)
```

```
fig, ax = plt.subplots(2, figsize=(10, 9))
ax[0].scatter(x = df['Sales'], y = df['Profit'],color = "blue", edgecolors = "white", linewidths = 0.1, alpha = 0.7)
ax[0].set_xlabel("Sales")
ax[0].set_ylabel("Profit")


ax[1].scatter(x = df['Quantity'], y = df['Profit'])
ax[1].set_xlabel("Quantity")
ax[1].set_ylabel("Price")

plt.show()
```

```
#Sales columns

figure = plt.figure(figsize=(15,8))
plt.hist(df['Sales'],color = 'olive',bins = 30)
plt.xlabel('Sales')
plt.ylabel('Total of States')
plt.title('Histogram of States')
```

```
#Profit Columns

figure = plt.figure(figsize=(15,8))
plt.hist(df['Profit'],color = 'olive',bins = 30)
plt.xlabel('Profit')
plt.ylabel('Total of Profit')
plt.title('Histogram of Profit')
```

```
# Plot Heat Map

#create the correlation matrix heat map
plt.figure(figsize=(14,12))

sns.heatmap(df.corr(),linewidths=.1, cmap="YlGnBu", annot=True)

plt.yticks(rotation=0);
```
